```python
import os

from google.cloud import bigquery

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "../../../squadbase-demo.json"

client = bigquery.Client()
schema_query = """
    SELECT
        column_name       AS col_name,
        data_type         AS data_type,
        is_nullable       AS nullable,
        ordinal_position  AS position
    FROM `bigquery-public-data.google_analytics_sample.INFORMATION_SCHEMA.COLUMNS`
    WHERE table_name = 'ga_sessions_20170801'
    ORDER BY ordinal_position;
"""
schema_result = client.query(schema_query)
schema_df = schema_result.to_dataframe()

schema_json = schema_df.to_json()

schema_json
```

```python
select_query = """
    SELECT
    t.*                      -- まずは全部
    EXCEPT(totals, trafficSource, device, geoNetwork)   -- ただし展開する構造体は除外
    ,  t.totals.*                               -- totals.visits, totals.hits … が列になる
    ,  t.trafficSource.*                        -- trafficSource.source, … が列になる
    ,  t.device.*
    ,  t.geoNetwork.*
    FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20170801`  AS t
    LIMIT 100;
"""
select_result = client.query(select_query)
select_result.to_dataframe()
```
